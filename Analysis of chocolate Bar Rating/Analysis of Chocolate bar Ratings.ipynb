```python
new_col_names=['company','bean_origin','ref','review_date','cocoa_per','company_location','rating','bean_type','broad_bean_origin']
df=df.rename(columns=dict(zip(df.columns,new_col_names)))
df.head()
```

```python
df['cocoa_per']=df['cocoa_per'].str.replace('%','').astype(float)/100
```

```python
df.head(15)
```

```python
unsatisfactory=df[df['rating']<3.0]
satisfactory=df[(df['rating']>=3.0) & (df['rating']<4.0)]
pre_elite=df[df['rating']>=4.0]

label_names=['Unsatisfactory','Satisfactory','Elite & Premium']
sizes=[unsatisfactory.shape[0],satisfactory.shape[0],pre_elite.shape[0]]

explode=(0.05,0.05,0.05)

plt.pie(sizes,labels=label_names,explode=explode,autopct='%1.1f%%',pctdistance=0.85,startangle=90,shadow=True)
fig=plt.gcf()

my_circle=plt.Circle((0,0),0.7,color='white')
fig.gca().add_artist(my_circle)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='rating',data=df)
plt.xlabel('Rating of chocolate bar',size=12,color='red')
plt.ylabel('Number of chocolate bar',size=12,color='red')
plt.show()
```

# Is there a relation between ‘Cocoa Percent’ and ‘Rating’ ?

```python
sns.lmplot(x='cocoa_per',y='rating',fit_reg=False,scatter_kws={"color":"darkred","alpha":0.3,"s":100},data=df)
plt.xlabel("Percentage of Cocoa",size=12,color="darkred")
plt.ylabel("Expert rating of the bars",size=12,color="darkred")
plt.show()
```

# Where are the best cocoa beans grown ?

```python
#to get the indices
countries=df['broad_bean_origin'].value_counts().index.tolist()[:5]
```

```python
best_choc={}
for j in countries:
    c=0
    b=df[df["broad_bean_origin"]==j]
    br=b[b['rating']>=3]
    for i in br['rating']:
        c+=1
        best_choc[j]=c
        
```

```python
best_choc
```

```python
li=best_choc.keys()
li
```

```python
plt.bar(range(len(best_choc)),best_choc.values(),align='center',color=['#a22a2a','#511515','#a22a2a','#d04949','#e59a9a'])
plt.xticks(range(len(best_choc)),list(li))
plt.xlabel("Country",color="brown",size=15)
plt.ylabel("Number of Chocolate bars",color="brown",size=15)
plt.show()
```

# Which countries produce the highest rated bars ?

```python
countries=df['company_location'].value_counts().index.tolist()
best_choc={}
for j in countries:
    c=0
    b=df[df['company_location']==j]
    br=b[b['rating']>=4]
    for i in br['rating']:
        
        c+=1
        best_choc[j]=c
li=best_choc.keys()
plt.hlines(y=li,xmin=0,xmax=best_choc.values(),color="darkgreen")
plt.plot(best_choc.values(),li,"o")
plt.xlabel('Country',color='brown',size=15)
plt.ylabel('Number of chocolate bars',color='brown',size=15)
plt.show()
```
